```python
import numpy as np
```

```python
# Subject in column(3)
# Student in row(5)
student_marks = np.array([
    [92, 65, 87],
    [99, 98, 92],
    [69, 66, 67],
    [77, 42, 99],
    [99, 33, 44]
])
student_marks
```

```python
print(f"Dimention: {student_marks.ndim}")
print(f"Size: {student_marks.size}")
print(f"Shape: {student_marks.shape}")
```

```python
# Average marks per student
average_marks = np.mean(student_marks, axis=1)
print("Average marks per student:", average_marks)
```

```python
# Subject-wise highest & lowest marks
high_marks = np.max(student_marks, axis=0)
print(f"Highest Marks: {high_marks}")

low_marks = np.min(student_marks, axis=0)
print(f"Lowest Marks: {low_marks}")
```

```python
# Overall topper
topper = np.max(np.mean(student_marks, axis=1))
print(topper)
```

```python
# Overall topper
average_marks = np.mean(student_marks, axis=1)
topper_index = np.argmax(average_marks)  # Index of the student with the highest average
topper_score = average_marks[topper_index]  # Highest average marks
print(f"Overall Topper: Student {topper_index + 1} with average marks: {topper_score}")
```

```python
# Assign grades based on average marks
def assign_grade(avg):
    if avg >= 90:
        return 'A'
    elif avg >= 75:
        return 'B'
    elif avg >= 60:
        return 'C'
    else:
        return 'D'

grades = []
for avg in average_marks:
    grades.append(assign_grade(avg))

print("Grades for all students:", grades)
```
