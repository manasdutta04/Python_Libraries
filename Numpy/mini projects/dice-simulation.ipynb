### Dice Simulation

**What to build**

- Simulate rolling dice 10,000 times
- Calculate probability of outcomes

**NumPy concepts**

- `np.random`
- Vectorized operations
- Frequency counting

```python
import numpy as np
```

```python
# Simulate rolling a die 10,000 times
rolls = np.random.randint(1, 7, 10000)
print(f"Total rolls: {len(rolls)}")
print(rolls)
```

```python
# Calculate frequencies of each outcome (1-6)
frequencies = np.bincount(rolls)[1:]
print(f"Frequencies: {frequencies}")

# Calculate probabilities
probabilities = frequencies / len(rolls)
print(f"Probabilities: {probabilities}")

for i in range(6):
    print(f"Outcome {i+1}: Probability = {probabilities[i]}")
```
